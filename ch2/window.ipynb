```python
import pyautogui as p
```

```python
# 마우스 좌표 가져오기
pos = p.position()
print(pos.x, pos.y)
```

```python
# 마우스 이동
# moveTo(x좌표, y좌표, 시간)
p.moveTo(100,100,0.5)
p.moveTo(200,200,0.5)
p.moveTo(300,300,0.5)
```

```python
# move() : 현재위치에서 얼마 갈것이지 => 상대좌표
# moveTo() : 절대좌표
p.moveTo(300,300,0.5)
p.move(100,100,0,25)
print(p.position())
p.move(100,100,0.25)
```

```python
p.sleep(3)
print(p.position())
```

```python
# 마우스 클릭

p.click(62,20,1)
```

```python
p.sleep(1)
# 클릭 횟수
p.click(clicks=5)
```

```python
p.click(clicks=2, interval=2, button="right")
```

```python
p.doubleClick(487,142)
```

```python
p.rightClick()
```

```python
p.mouseInfo()
```

```python
p.sleep(3)
p.moveTo(700,600)
p.mouseDown()
p.moveTo(1000,1000)
p.mouseUp()
```

```python
p.sleep(2)
p.moveTo(854,220)
p.drag(100,0,0.25)
```

```python
# 양수 : 위로, 음수 : 아래로
p.sleep(3)
p.scroll(-3000)
```
